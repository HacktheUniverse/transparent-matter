```python
import os,sys
import numpy as np
import pylab as pl
import scipy.constants as cons
%pylab inline
from matplotlib import rcParams
rcParams["savefig.dpi"] = 150
kpc2ly=3261.63344
ly2m=1.0/1.05702341e-16

kpc2m=3e19
```

```python
def newton(M,R):
    return np.sqrt(cons.G*M/R/ly2m)
```

```python
def iso(Rc,Vinf,R):
    tmp=1.0-Rc/R/ly2m*np.arctan(R*ly2m/Rc)
    return Vinf*np.sqrt(tmp)
```

```python
#finite radii
gal='M33'
M={'M33':1e40, 'Vinf':105400, 'Rc':1.39*kpc2m}
R=np.arange(1,40)*1000.
model={}

model['nodm']={'v':newton(M[gal],R)}

model['ISO']={'v':iso(M['Rc'],M['Vinf'],R)}
```

```python
pl.plot(R,model['nodm']['v'])
```

```python
pl.plot(R,model['ISO']['v'])
```
